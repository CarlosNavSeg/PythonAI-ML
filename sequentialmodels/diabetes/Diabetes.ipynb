```python
df = pd.read_csv('/content/drive/MyDrive/PIA23-24/tema3/diabetes/diabetes.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = df.drop('Outcome', 1)
y = df['Outcome']
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras import Sequential, layers

X_test, X_train, y_test, y_train = train_test_split(X, y, test_size = 0.2)
X_val, X_test, y_val, y_test = train_test_split(X,y ,test_size = 0.5)
```

```python
model = tf.keras.models.Sequential([
      layers.Dense(16, activation="relu", name="layer1"),
      layers.Dense(32, activation="relu", name="layer2"),
      layers.Dense(2, name="layer3", activation="softmax"),
  ])

    # Compila el modelo
model.compile(loss="sparse_categorical_crossentropy", optimizer = "adam", metrics=['accuracy'])# Entrena el modelo
historico = model.fit(X_train, y_train, epochs=100, validation_data=(X_val, y_val), batch_size = 16)
```

```python
y_pred = pd.DataFrame(model.predict(X_test), columns=['No tiene diabetes', 'Tiene'])
```

```python
model.evaluate(X_test, y_test)
```
